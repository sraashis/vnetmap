```python
import os
import networkx as nx
import matplotlib.pyplot as plt
import random
import string
import numpy as np
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
PARAMS = {
    'NODE_TYPES': ['ROUTER', 'SWITCH', 'CACHE', 'FIREWALL', 'LOAD_BALANCER'],
    'NODE_CAPACITIES': np.arange(20, 200, 20),
    'LINK_CAPACITIES': np.arange(20, 100, 5)
}
VPARAMS = {
    'NODE_TYPES': ['ROUTER', 'SWITCH', 'CACHE', 'FIREWALL', 'LOAD_BALANCER'],
    'NODE_CAPACITIES': np.arange(20, 40, 5),
    'LINK_CAPACITIES': np.arange(20, 40, 5)
}
```

```python
NODE_COLOR = {
    'ROUTER': 'GREEN',
    'SWITCH': 'LIGHTBLUE',
    'CACHE': 'BROWN',
    'FIREWALL': 'RED',
    'LOAD_BALANCER': 'YELLOW'
}
```

### Generate graph

```python
def gen_network(num_nodes=None, num_links=None, root_dir='networks',\
                net_file=None, override=False, params=None): 
    
    full_path = root_dir + os.sep + net_file
    if os.path.isfile(full_path) and not override:
        print('### Saved graph loaded: ' + full_path)
        return nx.read_graphml(full_path)
    
    net = nx.Graph()
    nodes = [n for n in string.ascii_uppercase[:num_nodes]]
        
    while num_links>0:
        n1 = random.choice(nodes)
        n2 = random.choice(nodes)
        if not net.has_edge(n1, n2):
            net.add_edge(n1, n2)
            num_links -= 1
            
    for n in net.nodes:
        net.nodes[n]['NODE_TYPE'] = random.choice(params['NODE_TYPES'])
        net.nodes[n]['NODE_CAPACITY'] = random.choice(params['NODE_CAPACITIES'])
    
    for e in net.edges:
        net.edges[e]['LINK_CAPACITY'] = random.choice(params['LINK_CAPACITIES'])
    print('### New graph loaded: ' + full_path)
    nx.write_graphml(net, full_path)
    return net
```

```python
def get_pos(net=None, root_dir='networks', net_file=None, override=False):
    pos_file = root_dir+os.sep+'pos_'+net_file.split('.')[0]+'.json'
    if os.path.isfile(pos_file) and not override:
        print('### Saved pos loaded: ' + pos_file)
        return json.load(open(pos_file, 'r'))
    
    pos = nx.spring_layout(net)
    for k in pos:
        pos[k] = list(pos[k])
    with open(pos_file, 'w') as outfile:
        json.dump(pos, outfile)
    print('### New pos loaded: ' + pos_file)
    return pos
```

```python
saved_net = 'net001.gml'
net = gen_network(num_nodes=10, num_links=22, \
                  net_file=saved_net, override=False, params=PARAMS)
pos = get_pos(net=net, net_file=saved_net, override=False)
```

```python
plt.rcParams['figure.figsize'] = 8, 8
plt.rcParams.update({'font.size': 8})
plt.rcParams.update({'font.family': 'console'})

####### Node colors based on type ##########
nodecolors = {}
for n in net.nodes:
    nodecolors[n] = NODE_COLOR[net.nodes[n]['NODE_TYPE']]

####### LINK CAPACITY as edge weights ##########
edge_lbls = {}
for e in net.edges:    
    edge_lbls[e] = net.edges[e]['LINK_CAPACITY']
    
####### LINK CAPACITY as edge width ##########
weights = [net[u][v]['LINK_CAPACITY'] for u,v in list(net.edges)]
weights = np.array(weights)/max(weights) * 2

nx.draw(net, node_color=nodecolors.values(),pos=pos, with_labels=True,\
        width=weights, edgecolors='blue')
nx.draw_networkx_edge_labels(net, pos, edge_labels=edge_lbls)
for node, (x, y) in pos.items():
    plt.text(x, y+0.055, \
             s=net.nodes[node]['NODE_TYPE']+' cap:'+\
             str(net.nodes[node]['NODE_CAPACITY']),\
             bbox=dict(facecolor='brown', alpha=0.2),\
             horizontalalignment='center')
plt.show()
```
